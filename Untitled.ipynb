```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np

# Load the moon map image
img = mpimg.imread('images/phobos.jpg')  # Replace with your path to the moon map

# Create a figure and axis
fig, ax = plt.subplots()

# Show the moon map image
ax.imshow(img, extent=[-180, 180, -90, 90],cmap = 'Greys')  # Set the extent to match coordinates (Longitude, Latitude)

# Add grid
for lon in np.linspace(-180, 180, 20):  # Longitude grid lines
    ax.plot([lon, lon], [-90, 90], color='white', linestyle='-', linewidth=0.5)

# Manually plot horizontal grid lines (latitude lines)
for lat in np.linspace(-90, 90, 20):  # Latitude grid lines
    ax.plot([-180, 180], [lat, lat], color='white', linestyle='-', linewidth=0.5)

# Set the grid's latitude and longitude ticks (adjust as needed)
ax.set_xticks(np.linspace(-180, 180, 13))  # Longitude grid lines
ax.set_yticks(np.linspace(-90, 90, 7))    # Latitude grid lines

ax.set_axis_off()
ax.set_xticks([]) 
ax.set_yticks([])
plt.tight_layout()
plt.savefig('images/phobos_with_grid.jpg', dpi=300, bbox_inches='tight',pad_inches=0)

# Show the plot with grid
plt.show()


```
